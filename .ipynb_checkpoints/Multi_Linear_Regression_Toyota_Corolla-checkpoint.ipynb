# 4. Model Building and Training

```python
linear_model_1=smf.ols('Price~Age_08_04', data=toyata_data).fit()

print('R-squared      :',linear_model_1.rsquared.round(4))
print('Adj R-squared  :',linear_model_1.rsquared_adj.round(4))
print('AIC Value      :',linear_model_1.aic.round(4))
print('BIC Value      :',linear_model_1.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_2=smf.ols('Price~Age_08_04+KM', data=toyata_data).fit()

print('R-squared      :',linear_model_2.rsquared.round(4))
print('Adj R-squared  :',linear_model_2.rsquared_adj.round(4))
print('AIC Value      :',linear_model_2.aic.round(4))
print('BIC Value      :',linear_model_2.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_3=smf.ols('Price~Age_08_04+KM+HP', data=toyata_data).fit()

print('R-squared      :',linear_model_3.rsquared.round(4))
print('Adj R-squared  :',linear_model_3.rsquared_adj.round(4))
print('AIC Value      :',linear_model_3.aic.round(4))
print('BIC Value      :',linear_model_3.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_4=smf.ols('Price~Age_08_04+KM+HP+cc', data=toyata_data).fit()

print('R-squared      :',linear_model_4.rsquared.round(4))
print('Adj R-squared  :',linear_model_4.rsquared_adj.round(4))
print('AIC Value      :',linear_model_4.aic.round(4))
print('BIC Value      :',linear_model_4.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_5=smf.ols('Price~Age_08_04+KM+HP+cc+Doors', data=toyata_data).fit()

print('R-squared      :',linear_model_5.rsquared.round(4))
print('Adj R-squared  :',linear_model_5.rsquared_adj.round(4))
print('AIC Value      :',linear_model_5.aic.round(4))
print('BIC Value      :',linear_model_5.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_6=smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears', data=toyata_data).fit()

print('R-squared      :',linear_model_6.rsquared.round(4))
print('Adj R-squared  :',linear_model_6.rsquared_adj.round(4))
print('AIC Value      :',linear_model_6.aic.round(4))
print('BIC Value      :',linear_model_6.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_7=smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax', data=toyata_data).fit()

print('R-squared      :',linear_model_7.rsquared.round(4))
print('Adj R-squared  :',linear_model_7.rsquared_adj.round(4))
print('AIC Value      :',linear_model_7.aic.round(4))
print('BIC Value      :',linear_model_7.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_8=smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight', data=toyata_data).fit()

print('R-squared      :',linear_model_8.rsquared.round(4))
print('Adj R-squared  :',linear_model_8.rsquared_adj.round(4))
print('AIC Value      :',linear_model_8.aic.round(4))
print('BIC Value      :',linear_model_8.bic.round(4))

#linear_model_1.pvalues
```

```python
linear_model_9=smf.ols('Price~Age_08_04+KM+HP+Quarterly_Tax+Weight', data=toyata_data).fit()

print('R-squared      :',linear_model_9.rsquared.round(4))
print('Adj R-squared  :',linear_model_9.rsquared_adj.round(4))
print('AIC Value      :',linear_model_9.aic.round(4))
print('BIC Value      :',linear_model_9.bic.round(4))

#linear_model_1.pvalues
```

# From all the 9 models we can see that linear_model_9 gives :
* High R-squared and R-squared-adj values
* Least AIC and BIC values
### cc, Doors and Gears does not play major role in prediction due to less correlation with the output. Hence, those features are removed to make the prediction model less complex
